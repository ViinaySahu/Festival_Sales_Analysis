```python
# importing required libraries

import numpy as np               # For Array
import pandas as pd              # For Data Frame
import matplotlib.pyplot as plt  # For Visualizing Data
%matplotlib inline
import seaborn as sns            # For Visualizing Data
```

```python
# importing csv file for analysis

df = pd.read_csv('Festival_Sales_Data.csv', encoding= 'unicode_escape')
# using 'unicode_escape' to avoid encoding error
```

```python
# finding number of rows and columns

df.shape
```

```python
# printing first 5 rows of file

df.head()
```

```python
# information of each columns

df.info()
```

```python
# columns 'Status' and 'unnamed1' has no values so we have to drop them.

df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
# checking and finding number of null values

pd.isnull(df).sum()
```

```python
# column 'Amount' has 12 null values so we have to drop these null values

df.dropna(inplace=True)
```

```python
# data type of column 'Amount' is float which we want to change in integer

df['Amount'] = df['Amount'].astype('int')
```

```python
# to see some stats

df.describe()
```

# Exploratory Data Analysis

### Gender

```python
# plotting a bar chart for Gender with count

ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:           # for count
    ax.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

### *From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

## Age

```python
# finding age group with high purchasing power

ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

## State

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
# total amount from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

#### *From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


## Marital Status

```python
# number of married and unmarried customer
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(10,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# distinguis between genders

sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

#### *From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

# Occupation

```python
# finding sectors with high order count

sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# finding sectors with higher order amount
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
```

#### *From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

# Product Category

```python
# product category with number of orders
sns.set(rc={'figure.figsize':(25,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# product category with higher order amount
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

#### *From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

## Top 10 most sold products

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```

# Conclusion:

### *Married women age group 26-35 yrs from UP,  Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category.*
